# Define the scrapping funcion

```python
def extract_images_path(url):
    
    request = requests.get(url)

    soup = BeautifulSoup(request.content, 'html.parser')
    
    image_urls = []
    for image in soup.find("div", {"class": "container-chapter-reader"}).findAll('img'):
        image_urls.append(image["src"])
    return image_urls
```

```python
def extract_images_path_option2(url):
    
    request = requests.get(url)

    soup = BeautifulSoup(request.content, 'html.parser')
    
    image_urls = []
    for image in soup.find_all("div", {"class": "page-break no-gaps"}):
        image_urls.append(image.find('img')["src"].replace("\t", "").replace("\n", ""))
    return image_urls
```

```python
def extract_images_path_option3(url):
    
    request = requests.get(url)

    soup = BeautifulSoup(request.content, 'html.parser')
    soup.find("div", {"class": "code-block code-block-21"}).decompose()
    
    image_urls = []
    #soup.find("div", {"class": "code-block code-block-21"}).decompose()
    for image in soup.find_all("div", {"class": "page-break "}):
        image_urls.append(image.find("img")["data-src"].replace(" \n ", ""))
    return image_urls
```

# readmanganato

```python
initial_chapter = 4
final_chapter = 60

base_url = "https://readmanganato.com/manga-mg989389/"
saving_path = r'Z:\JD\Story Of Low Rank Soldier Becoming A Monarch'

for chapter in range(initial_chapter, final_chapter+1):
    chapter_url = base_url + "chapter-{}".format(chapter)
    print(chapter_url)
    images_paths = extract_images_path(chapter_url)
    with requests.Session() as session:
        for p in images_paths:
            print(p)
            resp_2 = session.get(p,
                                 headers={"referer":dirname(p)},
                                 verify=False) # https://stackoverflow.com/questions/10667960/python-requests-throwing-sslerror
            img_path_zero_fill = "00" if len(basename(p).split(".")[0])== 1 else "0"
            chapter_path_zero_fill = "00" if len(str(chapter))== 1 else "0" if len(str(chapter))== 2 else ""
            with open(r"{}\chapter_{}{}_{}{}".format(saving_path, chapter_path_zero_fill, chapter, img_path_zero_fill, basename(p)), 'wb') as outfile:
                outfile.write(resp_2.content)
            sleep(1.5)
    print("chapter {} finished".format(chapter))
    sleep(4)

print("THE END!")
```
