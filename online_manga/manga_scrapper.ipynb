```python
from selenium import webdriver

def get_soup_html_with_browser(url_page):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_argument('--profile-directory=Default')
    chrome_options.add_argument("--incognito")
    chrome_options.add_argument("--disable-plugins-discovery")
    chrome_options.add_argument("--start-maximized")
    #chrome_options.add_argument("--headless") # do not open browser

    driver = webdriver.Chrome(options=chrome_options)
    print("Opening browser")
    driver.get(url_page)
    sleep(8)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    sleep(3)
    driver.quit()
    print("chrome closed")
    return soup
```

```python
def extract_images_path(url):
    
    request = requests.get(url)

    soup = BeautifulSoup(request.content, 'html.parser')
    
    image_urls = []
    for image in soup.find("div", {"class": "container-chapter-reader"}).findAll('img'):
        image_urls.append(image["src"])
    return image_urls
```

```python
def extract_images_path_option2(url):
    
    request = requests.get(url)

    soup = BeautifulSoup(request.content, 'html.parser')
    
    image_urls = []
    for image in soup.find_all("div", {"class": "page-break no-gaps"}):
        image_urls.append(image.find('img')["src"].replace("\t", "").replace("\n", ""))
    return image_urls
```

```python
def extract_images_path_option3(url):
    image_urls = []
    request = requests.get(url)
    if request.status_code==403:
        sleep(2)
        soup = get_soup_html_with_browser(url)

        for image in soup.find("div", {"id": "readerarea"}).find_all('img'):
            image_urls.append(image["src"])
    else:
        soup = BeautifulSoup(request.content, 'html.parser')
        soup.find("div", {"class": "code-block code-block-21"}).decompose()
        for image in soup.find_all("div", {"class": "page-break "}):
            image_urls.append(image.find("img")["data-src"].replace(" \n ", ""))

    return image_urls
```

```python
# for the testing cloud function

import functions_framework

from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from bs4 import BeautifulSoup

from google.cloud import storage
import requests
from os.path import dirname, basename

def get_soup_html_with_browser(url_page):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--window-size=1920,1080')
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--profile-directory=Default')
    chrome_options.add_argument("--incognito")
    chrome_options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options=chrome_options)
    print("Opening browser")
    driver.get(url_page)
    sleep(5)

    driver.find_element(by=By.XPATH, value="/html/body/section/div[1]/div[2]/div[1]/div/select/option[2]").click()

    #select_element = driver.find_element(By.CLASS, 'number')
    #select = Select(select_element)
    #select.select_by_value('1')

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    sleep(3)
    driver.quit()
    print("chrome closed")
    return soup


@functions_framework.http
def hello_http(request):

    request_json = request.get_json(silent=True)
    print(request_json)


    initial_chapter = request_json["start"]
    final_chapter = request_json["end"]

    base_url = "https://www.leercapitulo.co/leer/dkk8d3/auto-hunting/"

    for chapter in range(initial_chapter, final_chapter+1):
        chapter_url = base_url + "{}/".format(chapter)
        print(chapter_url)
        soup = get_soup_html_with_browser(chapter_url)


        url_images = []
        for image in soup.find("div", {"class": "comic_wraCon text-center"}).findAll('img'):
            url_images.append(image["src"])

        with requests.Session() as session:
            for n, p in enumerate(url_images):

                resp_2 = session.get(p)
                img_path_zero_fill = "00" if len(basename(p).split(".")[0])== 1 else "0"
                chapter_path_zero_fill = "000" if len(str(chapter))== 1 else "00" if len(str(chapter))== 2 else "0"

                storage_client = storage.Client()
                bucket = storage_client.bucket("scraping-test-0012")

                blob = bucket.blob(r"chapter_{}{}_{}{:02}.jpg".format(chapter_path_zero_fill, chapter, img_path_zero_fill, n+1))

                with blob.open("wb") as f:
                    f.write(resp_2.content)
                sleep(0.5)


        print({"Chapter {}".format(chapter):url_images})
        #print({"chapter {} finished".format(chapter): images_paths })
        sleep(2)

    print("THE END!")

    return "Finish"
```
